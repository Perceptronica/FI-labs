```python
import numpy as np 
from matplotlib import pyplot as plt 
```

```python
plt.grid()
rectangle = plt.Rectangle((5, -5), 10, -10, fc="white", ec="red")
plt.gca().add_patch(rectangle)
plt.axis([-20, 20, -20, 20])
#plt.axis('scaled')
i, j, l = 11, 13, 10
plt.plot(i, j, 'bo')
for k in range(1, 51):
    i = abs(k - 15) - min(i/3, (j + l) % 10) -20
    j = -(j + k)/5 + abs(i*l % 8)
    l = max((i + k) % 15, (l + k) % 14)
    plt.plot(i, j, 'go')
plt.plot(i, j, 'ro')
plt.show()
```
